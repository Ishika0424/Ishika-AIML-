```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Dataset Import
# Load the dataset (e.g., Iris or Titanic)
dataset = pd.read_csv('supermarket_sales.csv')

# Display dataset structure
print("Dataset Structure:")
print(dataset.info())

# Display the first few rows
print("First 5 rows:")
print(dataset.head())

# 2. Basic Visualizations (Matplotlib)
# Histogram
plt.figure(figsize=(8, 6))
dataset['Total'].hist(bins=20, color='blue', alpha=0.7)
plt.title('Histogram of Total')
plt.xlabel('Numerical Column')
plt.ylabel('Frequency')
plt.show()

# Bar Plot (For categorical data)
plt.figure(figsize=(8, 6))
dataset['Total'].value_counts().plot(kind='bar', color='green')
plt.title('Bar Plot of Categorical_Column')
plt.xlabel('Categories')
plt.ylabel('Frequency')
plt.show()

# Pie Chart (For categorical data distribution)
category_counts = dataset['Total'].value_counts()
plt.figure(figsize=(8, 6))
category_counts.plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('Pie Chart of Categorical_Column')
plt.ylabel('')  # Hide y-label
plt.show()

# 3. Advanced Visualizations (Seaborn)
# Box Plot (For numerical data analysis)
plt.figure(figsize=(8, 6))
sns.boxplot(x='Total', y='Quantity', data=dataset)
plt.title('Box Plot: Categorical_Column vs Numerical_Column')
plt.show()

# Violin Plot (For distribution across categories)
plt.figure(figsize=(8, 6))
sns.violinplot(x='Total', y='Quantity', data=dataset)
plt.title('Violin Plot: Categorical_Column vs Numerical_Column')
plt.show()

# Pair Plot (For pairwise relationships)
sns.pairplot(dataset[['Total', 'Date', 'Time']])  # Select columns to plot
plt.title('Pair Plot')
plt.show()


# Select only numerical columns for correlation
numerical_cols = dataset.select_dtypes(include=['number']).columns
# Calculate correlation matrix
correlation_matrix = dataset[numerical_cols].corr()
# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()

# 4. Customizations (e.g., Titles, Labels)
# Example customization for histogram
plt.figure(figsize=(8, 6))
dataset['Total'].hist(bins=20, color='purple', alpha=0.7)
plt.title('Customized Histogram of Column_Name', fontsize=16)
plt.xlabel('Column_Name', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.show()

# 5. Conclusion (Analysis)
# Based on the plots, discuss trends, outliers, correlations, and any significant findings.
```
