```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import skew
import numpy as np
```

```python
data = pd.read_csv("Real estate.csv")
```

```python
data.info()
```

```python
 # Step 4: Check for missing data
print("\nColumns with missing data:")
missing_data = data.isnull().sum()
print(missing_data[missing_data > 0])
```

```python
 # Step 5: Check for inconsistent data in categorical columns
print("\nUnique values in each categorical column (to check for inconsistencies):")
for col in data.select_dtypes(include=['object']).columns:
    print(f"{col}: {data[col].unique()}")
```

```python
print("Statistical Summary:")
summary = data.describe()  # Generates mean, std, min, max, 25th, 50th (median), and 75th percentiles
print(summary)

# Step 3: Include the median separately for better insights
print("\nMedian for numerical columns:")
print(data.median(numeric_only=True))  # Median isn't included in describe()
```

```python
print("Missing Values:")
missing_data = data.isnull().sum()  # Count of missing values for each column
print(missing_data[missing_data > 0])  # Display only columns with missing values

# Step 3: Check for duplicate rows
print("\nDuplicate Rows:")
duplicate_rows = data.duplicated().sum()  # Count of duplicate rows
print(f"Number of duplicate rows: {duplicate_rows}")

# Step 4: Handle missing values
# Option 1: Remove rows with missing values (if they are minimal)
df_cleaned = data.dropna()
print("\nRows after removing missing values:")
print(df_cleaned.info())

# Option 2: Fill missing values (for numeric columns, for example)
df_filled = data.fillna(data.mean())  # Fill missing numeric values with the column mean
print("\nRows after filling missing values:")
print(df_filled.info())

# Step 5: Handle duplicate values
# Option 1: Remove duplicate rows
df_no_duplicates = data.drop_duplicates()
print("\nRows after removing duplicates:")
print(df_no_duplicates.info())
```

```python
# Plot histogram for 'Y house price of unit area'
plt.figure(figsize=(8, 5))
plt.hist(data['Y house price of unit area'], bins=30, edgecolor='black', alpha=0.7)
plt.title('Distribution of House Prices')
plt.xlabel('House Prices')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

# Check skewness
from scipy.stats import skew
skewness = skew(data['Y house price of unit area'])
print(f"Skewness of SalePrice: {skewness}")

# If skewed, apply log transformation
if skewness > 0.5 or skewness < -0.5:
    data['Log_SalePrice'] = np.log1p(data['Y house price of unit area'])
    plt.figure(figsize=(8, 5))
    plt.hist(data['Log_SalePrice'], bins=30, edgecolor='black', alpha=0.7)
    plt.title('Log-Transformed Distribution of House Prices')
    plt.xlabel('Log(House Prices)')
    plt.ylabel('Frequency')
    plt.grid(True)
    plt.show()
```

```python
# Define the features (X1 to X6) and target (Y house price of unit area)
features = ['X1 transaction date', 'X2 house age', 'X3 distance to the nearest MRT station', 'X4 number of convenience stores', 'X5 latitude', 'X6 longitude']
target = 'Y house price of unit area'

# Plot scatter plots for each feature vsY house price of unit area
plt.figure(figsize=(20, 15))
for i, feature in enumerate(features, 1):
     plt.subplot(2, 3, i)  # Arrange plots in a grid (2 rows, 3 columns)
     plt.scatter(data[feature], data[target], alpha=0.7, edgecolors='black')
     plt.title(f'{feature} vs {target}')
     plt.xlabel(feature)
     plt.ylabel(target)
     plt.grid(True)

# Display the plots
plt.tight_layout()
plt.show()
```
